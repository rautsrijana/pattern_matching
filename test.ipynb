```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# For Shenzhen data
shenzhen_xrays_path = '/Users/srijanaraut/Desktop/research/Shenzhen/img'
shenzhen_masks_path = '/Users/srijanaraut/Desktop/research/Shenzhen/mask'
```

```python
# For Montgomery data
montgomery_xrays_path = '/Users/srijanaraut/Desktop/research/Montgomery/img'
montgomery_masks_path = '/Users/srijanaraut/Desktop/research/Montgomery/mask'
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
xray = load_images_from_folder(montgomery_xrays_path)
xray = cv2.cvtColor(xray[1], cv2.COLOR_RGB2GRAY)
plt.imshow(xray, cmap="gray")
```

```python
xray.shape
```

```python
mask = load_images_from_folder(montgomery_masks_path)
mask = cv2.cvtColor(mask[1], cv2.COLOR_RGB2GRAY)
plt.imshow(mask, cmap="gray")
```

```python
def apply_mask(image, mask):
    # Convert to numpy arrays
    image = np.array(image)
    mask = np.array(mask)
    
    mask = np.stack(mask)
    # Multiply arrays
    resultant = image*mask
    
    return resultant
```

```python
resultant = apply_mask(xray, mask)
```
